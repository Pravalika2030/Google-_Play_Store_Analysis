```python
# Top 5 Best Rated Social Media Apps:
top_social = social_df.sort_values(by=['Rating', 'Reviews'], ascending=[False, False])
top_5_social = top_social[['App', 'Rating', 'Reviews']].head(5)

print("🎯 Top 5 Best Rated Social Media Apps:")
print(top_5_social)
```

```python
#Bottom 5 Worst Rated Social Media Apps
worst_social = social_df.sort_values(by=['Rating', 'Reviews'], ascending=[True, True])
bottom_5_social = worst_social[['App', 'Rating', 'Reviews']].head(5)

print("\n💣 Bottom 5 Worst Rated Social Media Apps:")
print(bottom_5_social)
```

```python
# Box Plot of Ratings by Category
plt.figure(figsize=(14, 6))
sns.boxplot(x='Category', y='Rating', data=df)
plt.xticks(rotation=90)
plt.title("Ratings Distribution by Category")
plt.show()
```

```python
#Compare Best vs Worst Apps (Bar Chart)
top_5_social['Type'] = 'Best'
bottom_5_social['Type'] = 'Worst'

# Combine both
combined = pd.concat([top_5_social, bottom_5_social])

# Plot with custom colors
plt.figure(figsize=(12, 6))
sns.barplot(
    x='Rating',
    y='App',
    data=combined,
    hue='Type',
    palette={'Best': 'green', 'Worst': 'red'}
)
plt.title("Top vs Bottom Rated Social Media Apps")
plt.xlabel("Rating")
plt.ylabel("App")
plt.legend(title='App Type')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Drop rows with missing values in Rating and Category
df_cleaned = df[['App', 'Category', 'Rating', 'Installs', 'Size']].dropna()

# Get the highest-rated app in each category
top_rated_per_category = df_cleaned.loc[df_cleaned.groupby('Category')['Rating'].idxmax()]

# Seaborn relplot
sns.relplot(
    data=top_rated_per_category,
    x='Category',
    y='Rating',
    hue='Installs',
    size='Size',
    sizes=(50, 300),
    palette='mako',
    kind='scatter',
    height=6,
    aspect=2
)

plt.xticks(rotation=90)
plt.title("Highest Rated App per Category (RelPlot)")
plt.xlabel("App Category")
plt.ylabel("App Rating")
plt.tight_layout()
plt.show()

```
